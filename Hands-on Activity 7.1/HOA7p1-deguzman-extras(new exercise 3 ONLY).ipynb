```python
# 1. Using web scraping, search for the list of the hospitals, their address and contact information. Save the list in a new csv file, hospitals.csv.

import requests

# requests the url first that was retrieved from Centers for Medicare & Medicaid Services.
# source : https://data.cms.gov/provider-data/dataset/f4ga-b9gx#api:~:text=%22downloadURL%22%3A%20%22https%3A//data.cms.gov/provider%2Ddata/sites/default/files/resources/5462b19a756c53c1becccf13787d9157_1739210708/Promoting_Interoperability%2DHospital.csv%22%2C
# under metadata: dataset -> try it out -> execute -> response body : you can also download the .json but  I downloaded the one with .csv
dataset_url = requests.get("https://data.cms.gov/provider-data/sites/default/files/resources/5462b19a756c53c1becccf13787d9157_1739210708/Promoting_Interoperability-Hospital.csv")

if dataset_url.ok: # this is to check if the there exist a possible entry for .csvs from my understanding
    with open("datasets/Hospital.csv", "wb") as file: # this just saves the contents of the files url in a .csv file using a write function. ( 'wb' write-binary, similar to 'w' write but it is for online
                                                      # or a website... )
        file.write(dataset_url.content)
    print("Download complete, the .csv name is located at = 'datasets/Hospital.csv'")
else:
    print(f"Failed to retrieve data. Status code: {dataset_url.status_code}")
```

```python
import pandas as pd

df = pd.read_csv("datasets/Hospital.csv")
print(df.dtypes)
df.head(5)
```

```python
# 2. Using the generated hospitals.csv, convert the csv file into pandas dataframe. Prepare the data using the necessary preprocessing techniques.
# https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.astype.html

df['Meets criteria for promoting interoperability of EHRs'] = df['Meets criteria for promoting interoperability of EHRs'].astype('bool')
df['Start Date'] = pd.to_datetime(df['Start Date'])
df['End Date'] = pd.to_datetime(df['End Date'])
print(df.dtypes)
df.head(1)
```
