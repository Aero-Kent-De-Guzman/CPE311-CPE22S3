```python
import sys

class PassengerToShore():
    def __init__(self, contents):
        self.contents = contents

    def dict_sort(self):
        temp = list(self.contents.items())
        for i in range(len(temp)):
            for j in range(len(temp) - 1 - i):
                if temp[j][1] > temp[j + 1][1]:
                    temp[j], temp[j + 1] = temp[j + 1], temp[j]
        self.contents = dict(temp)

    def BoatCross(self):
        island = list(self.contents.items())
        boat = []
        shore = []
        temp = []
        boatKG = 0
        islandPosition = 0
        shorePosition = 0
        errorPromt = True
        supplyPromt = True

        print("##############################################################")
        print("\t  EVENT 1 : CARRYING SUPPLIES FIRST TO SHORE")
        print("##############################################################\n")
        print("current status : \nisland\t",island,"\nboat\t",boat,"\nshore\t",shore,"\n")
        for key, val in island:
            if key == 'Supplies':
                supplyPromt = False
                islandPosition = 0
                boatKG += val
                boat.append((key, val))
                island.pop(islandPosition)
                temp += island
                print("Supplies have boarded the boat\n\ncurrent status : \nisland\t",island,"\nboat\t",boat,"\nshore\t",shore,"\n")
                for key, val in island:
                    if (val + boat[0][1]) < 100:
                        boatKG += boat[0][1]
                        boat.append(island[islandPosition])
                        island.pop(islandPosition)
                        print("Person have boarded the boat\n\ncurrent status : \nisland\t",island,"\nboat\t",boat,"\nshore\t",shore,"\n")
                        islandPosition = 0
                        shore += boat
                        temp.clear()
                        temp += shore
                        boat.clear()
                        print("All were dropped off to shore\n\ncurrent status : \nisland\t",island,"\nboat\t",boat,"\nshore\t",shore,"\n")
                        boat.append(shore[1])
                        shore.pop(1)
                        print("Driver returning to island\n\ncurrent status : \nisland\t",island,"\nboat\t",boat,"\nshore\t",shore,"\n")
                        island = boat + island
                        boat.clear()
                        temp.clear()
                        print("current status : \nisland\t",island,"\nboat\t",boat,"\nshore\t",shore,"\n")
                        errorPromt = False
                        islandPosition = 0
                        break
                    islandPosition += 1
            islandPosition += 1

        if supplyPromt == True:
            sys.exit('ERROR! No Supplies detected! make sure the name is exactly "Supplies" case-sensitive.')
        if errorPromt == True:
            sys.exit('ERROR! Unable to carry the supplies will resort in a loop.')

        islandPosition = 0
        for key, val in island:
            islandPosition += 1
            if islandPosition < len(island):
                print(val + island[islandPosition][1])
                if val + island[islandPosition][1] <= 100:
                    break
                else:
                    errorPromt = True
        if errorPromt == True:
            sys.exit('ERROR! Having not a single pair cross together will result in a loop...')

        print("##############################################################")
        print(" EVENT 2 : Transfering the people in the island to the shore.")
        print("##############################################################\n")
        print("current status : \nisland\t",island,"\nboat\t",boat,"\nshore\t",shore,"\n")
        while(len(island) != 0):
            # For-loop to when going to the shore, as long as the boat can still carry, it will keep adding the lightest.
            print("______________________________________________________________\n\nCURRENTLY ON THE ISLAND\n\nPassengers are boarding...\n")
            boatKG = 0
            for key, val in island[:]:
                boatKG += val
                if boatKG <= 100:
                    boat.append((key,val))
                    temp = island.pop(0)
                    print("current status : \nisland\t",island,"\nboat\t",boat,"\nshore\t",shore,"\nboatKG\t",boatKG,'\n')
                else:
                    shore += boat
                    boat.clear()
                    boatKG = 0
                    print("Max capacity! going to shore while returning exceed to original position...\n\ncurrent status : \nisland\t",island,"\nboat\t",boat,"\nshore\t",shore,"\nboatKG\t",boatKG,'\n')
                    break
            if len(island) == 0:
                shore += boat
                boat.clear()
                print("Max capacity! Dropping everyone off to the shore...\n\ncurrent status : \nisland\t",island,"\nboat\t",boat,"\nshore\t",shore,"\nboatKG\t",boatKG,'\n')
                break
            # For-loop to when going back to the island, only one and the lightest will go back alone...
            print("______________________________________________________________\n\nCURRENTLY ON THE SHORE\n")
            shorePosition = 0
            for key, val in shore:
                if key == 'Supplies':
                    shorePosition += 1
                    continue
                boat.append((key, val))
                shore.pop(shorePosition)
                print("Driver from shore is entering the boat.\n\ncurrent status : \nisland\t",island,"\nboat\t",boat,"\nshore\t",shore,"\n")
                island += boat
                boat.clear()
                print("All people in the boat are dropping off.\n\ncurrent status : \nisland\t",island,"\nboat\t",boat,"\nshore\t",shore,"\n")
                break
        print("______________________________________________________________\n\nSUCCESSFULLY TRANSFERRED EVERYONE")
        print("current status : \nisland\t",island,"\nboat\t",boat,"\nshore\t",shore,"\n")

Passengers = {
    'Roman' : 90,
    'Verlyn' : 80,
   'Supplies' : 20, # Please do not change the name, it is the identifier used later on...
    'Lloyd' : 60,
    'Robin' : 40
    }

Boat = PassengerToShore(Passengers)
Boat.dict_sort()
Boat.BoatCross()
```
