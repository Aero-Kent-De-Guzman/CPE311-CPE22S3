```python
household    = data[['Country','Region','Household estimate (kg/capita/year)','Household estimate (tonnes/year)']]
retail       = data[['Country','Region','Retail estimate (kg/capita/year)','Retail estimate (tonnes/year)']]
food_service = data[['Country','Region','Food service estimate (kg/capita/year)','Food service estimate (tonnes/year)']]
```

```python
ASEAN = data[data['Region'] == 'South-eastern Asia'].copy()
ASEAN
```

```python
#Plot the top 10 countries with maximum combined estimate
ASEAN.nlargest(100, ['combined figures (kg/capita/year)']).plot(x="Country", y="combined figures (kg/capita/year)",
        kind="barh",
        figsize=(10, 8), 
        ylabel='Countries',
        xlabel='combined figures (kg/capita/year)',
        title='combined figures (kg/capita/year)')
```

```python
ASEAN.nlargest(100, ['Household estimate (kg/capita/year)']).plot(x="Country", y="Household estimate (kg/capita/year)",
        kind="barh",
        figsize=(10, 8), 
        ylabel='Countries',
        xlabel='Household estimate (kg/capita/year)',
        title='Household estimate (kg/capita/year)')
```

```python
ASEAN.nlargest(100, ['Retail estimate (kg/capita/year)']).plot(x="Country", y="Retail estimate (kg/capita/year)",
        kind="barh",
        figsize=(10, 8), 
        ylabel='Countries',
        xlabel='Retail estimate (kg/capita/year)',
        title='Retail estimate (kg/capita/year)')
```

```python
ASEAN.nlargest(100, ['Food service estimate (kg/capita/year)']).plot(x="Country", y="Food service estimate (kg/capita/year)",
        kind="barh",
        figsize=(10, 8), 
        ylabel='Countries',
        xlabel='Food service estimate (kg/capita/year)',
        title='Food service estimate (kg/capita/year)')
```

Based on the provided descriptive statistics detailing the visualization of the estimated food waste per sectors such as households, retail,
and food service it can be seen here that among all of the countries within the ASEAN group, Malaysia ranks at the top with more than 80
kg/capita/year in all of the sectors while the rest are usually less than 40 kg/capita/year in all sectors except household.

Looking at the pattern as well, it can be extracted here that most of the food waste that are being produced are actually being sourced mostly
from household due to most of the waste being generated are mostly and greatly in that specific sector.

Additionally, compared to Malaysia, most of the countries nearly has the same output in all sectors which suggests that while all countries
have slightly significat results in terms of their land size, the same cannot be said for Malaysia.

# *4. Perform correlation analysis.*

```python
ASEAN[[
    'combined figures (kg/capita/year)',
    'Household estimate (kg/capita/year)',
    'Retail estimate (kg/capita/year)',
    'Food service estimate (kg/capita/year)'
]].corr(min_periods=3)
```

# *5. Interpret the results based on the descriptive and correlation analysis.*

```python
Comparing the results from the previous descriptive statistics and the prior correlational, it can be seen here that there are very high
correlation in terms of the Food service and Retail sectors which suggest that most of the byproduct waste being produced are being sourced
from the same sources (e.g. they are sharing and providing each other their necessities which means that they share the waste themselves)
this can be further supported with the visualized output since when looking at the graphs for the three sectors, the Retail estimate
and Food service estimate nearly has the same results when compared to the Household sectors which has staggeringly more waste produced in
terms of kg/capita/year.
```
