## 1. Using the meteorite data from the Meteorite_Landings.csv file, create a pivot table that shows both the number of meteorites and the 95th percentile of meteorite mass for those that were found versus observed falling per year from 2005 through 2009 (inclusive). Hint: Be sure to convert the year column to a number as we did in the previous exercise.

```python
import pandas as pd

DF1 = pd.read_csv('datasets/Meteorite_Landings.csv')
DF1['year'] = DF1['year'].str[6:10]
DF1['year'] = pd.to_numeric(DF1['year'])
DF1.dtypes
```

```python
pTable = DF1[['year','mass (g)','fall']]
pTable.head(1)
```

```python
# reorganizing to only get between the given years
pTable = pTable[pTable.year >= 2005]
pTable = pTable[pTable.year <= 2009]
pTable = pTable.reset_index()
pTable
```

```python
pTable.describe(include = 'all') # does not contain a 95% quantile
```

```python
pTable.groupby('fall')[['year','mass (g)']].quantile(.95) # This now contains the 95 % quantile
                                                          # source : https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.Series.quantile.html
```

## 1. Using the taxi trip data in the 2019_Yellow_Taxi_Trip_Data.csv file, resample the data to an hourly frequency based on the dropoff time. Calculate the total trip_distance, fare_amount, tolls_amount, and tip_amount, then find the 5 hours with the most tips.

```python
import pandas as pd

DF3 = pd.read_csv('datasets/2019_Yellow_Taxi_Trip_Data.csv')
DF3.head(5)
```

```python
DF3.dtypes
```

```python
DF3[['tpep_pickup_datetime','tpep_dropoff_datetime']] = DF3[['tpep_pickup_datetime','tpep_dropoff_datetime']].apply(pd.to_datetime)
DF3.dtypes
```

```python
# source : https://pandas.pydata.org/docs/reference/api/pandas.Series.dt.hour.html
# converting to an hourly frequency...
DF3['tpep_pickup_datetime'] = DF3['tpep_pickup_datetime'].dt.hour
DF3['tpep_dropoff_datetime'] = DF3['tpep_dropoff_datetime'].dt.hour
```

```python
taxi = DF3[['tpep_pickup_datetime','tpep_dropoff_datetime','trip_distance','fare_amount','tolls_amount','tip_amount']]
taxi.head(1)
```

```python
taxi['hourly frequency'] = taxi['tpep_dropoff_datetime'] - taxi['tpep_pickup_datetime']
taxi['total tips']       = taxi['tip_amount'] - (taxi['fare_amount'] + taxi['tolls_amount'])
taxi = taxi.sort_values('total tips')
taxi_5_hours_highest = taxi[['hourly frequency','trip_distance','fare_amount','tolls_amount','tip_amount','total tips']].tail(5)
taxi_5_hours_highest # figuring out the hourly freq. based on the total tips which is the total from subtracting tip-amount to both fare and tolls amount
```
